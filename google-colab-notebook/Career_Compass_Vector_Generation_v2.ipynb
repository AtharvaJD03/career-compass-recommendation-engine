# Load Dataset

```python
import pandas as pd

df = pd.read_csv("career_dataset.csv")

print(df.head())
print("\nColumns:")
print(df.columns)
print("\nShape:", df.shape)
```

# Data Cleaning & Validation

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check data types
print("\nData types:")
print(df.dtypes)

# Check unique roles
print("\nUnique roles:")
print(df["role"].unique())
```

# Text Representation

```python
def create_text_representation(row):
    return (
        f"Skills: {row['skills']}. "
        f"Education: {row['education']}. "
        f"Interests: {row['interests']}. "
        f"Experience: {row['experience']} years."
    )

df["career_text"] = df.apply(create_text_representation, axis=1)

df[["career_text", "role"]].head()
```

#Apply TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["career_text"])

print("TF-IDF Shape:", tfidf_matrix.shape)
```

# Convert to JSON Vector Database

```python
import json

career_vectors = []

for i in range(len(df)):
    vector = tfidf_matrix[i].toarray()[0].tolist()
    job_role = df.iloc[i]["role"]

    career_vectors.append({
        "job_role": job_role,
        "vector": vector
    })

with open("career_vectors_v2.json", "w") as f:
    json.dump(career_vectors, f)

print("Vector database saved successfully!")
```

# Download JSON File

# Vocablury

```python
# Save vocabulary
vocab = vectorizer.vocabulary_

import json

with open("tfidf_vocabulary_v2.json", "w") as f:
    json.dump(vocab, f)

print("Vocabulary saved successfully!")
```

# Download Vocablury

```python
from google.colab import files
files.download("tfidf_vocabulary_v2.json")
```

# Download IDF Weights

```python
idf_values = vectorizer.idf_.tolist()

import json

with open("tfidf_idf_v2.json", "w") as f:
    json.dump(idf_values, f)

from google.colab import files
files.download("tfidf_idf_v2.json")
```
